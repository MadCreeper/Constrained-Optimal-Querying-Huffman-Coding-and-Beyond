```python
import seaborn as sns
import pandas as pd
from math import isclose
from matplotlib import pyplot as plt

import pickle

sns.set_theme()
```

```python
with open('experiment1.pickle', 'rb') as f:
    results = pickle.load(f)

gap = [(x[1] - x[0]) / x[0] for x in results]
data = pd.DataFrame({
    'gap': gap
})


a, b = zip(*results)

x_label = 'Expected length (brute force)'
y_label = 'Expected length (greedy Huffman)'

data2 = pd.DataFrame({
    x_label: a,
    y_label: b,
})
```

```python
print(sum([1 for x in gap if isclose(x, 0)]))
```

```python
g = sns.relplot(data=data2, x=x_label, y=y_label)
# g.ax.axline(xy1=(1.55, 1.55), slope=1, color="r", dashes=(5, 2))
g.ax.axline(xy1=(1.55, 1.55), slope=1, color="r")
plt.savefig('fig1.png', dpi=600, bbox_inches='tight')
```
